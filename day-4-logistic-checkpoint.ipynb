```python
titanic_data.drop('Cabin',axis=1,inplace=True)  #droping the entire column
```

```python
titanic_data.head()
```

```python
sex=pd.get_dummies(titanic_data['Sex'],drop_first=True) #drop_first=True  One column is enough. one hot encoding
sex
```

```python
embarked=pd.get_dummies(titanic_data['Embarked'],drop_first=True)    
embarked
```

```python
# remove the attribute which doesn't account to the final result. As ticket,passengerId doesnt give accurate result.
titanic_data=titanic_data.drop(['Embarked','Fare','Sex','Ticket','PassengerId','Title','Name'],axis=1)
```

```python
x=titanic_data.drop(['Survived'],axis=1) #inplce=True isn't present so the titanic_data is not updated by deleting the columns Survived
y=titanic_data.Survived  
y  #contains the output
```

```python
from sklearn.model_selection import train_test_split #cross_validation
from sklearn.linear_model import LogisticRegression
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2)  
#test_size=0.2 divides the data into 80-20 ratio.
# for training use 80% data and for testing use 20%data
```

```python

model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
pred=model.predict(X_test)
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
accuracy_score(y_test,pred)
```

```python
confusion_matrix(y_test,pred)
```
